```python
# densité de loi uniforme
import numpy as np
import matplotlib.pyplot as plt

def uniform_density(x, a, b):
    if a <= x <= b:
        density = 1 / (b - a)
    else:
        density = 0
    return density

# Exemple d'utilisation
a = 2
b = 5
c = 3.5

density = uniform_density(c, a, b)
print("Densité de probabilité :", density)

# Création des données pour l'axe x
x = np.linspace(a - 1, b + 1, 100)

# Calcul de la densité de probabilité pour chaque valeur de x
pdf = np.where((x >= a) & (x <= b), 1 / (b - a), 0)

# Tracé du graphique de la densité de probabilité
plt.plot(x, pdf, color='blue')
plt.title('Densité de la loi uniforme U(2, 5)')
plt.xlabel('x')
plt.ylabel('Densité de probabilité')
plt.grid(True)
plt.show()
```

```python
# La fonction de répartition de la loi unniforme 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Paramètres de la loi uniforme
a = 2  # Borne inférieure
b = 5  # Borne supérieure

# Création des données pour l'axe x
x = np.linspace(a - 1, b + 1, 100)

# Calcul de la fonction de répartition pour chaque valeur de x
cdf = uniform.cdf(x, loc=a, scale=b-a)

# Tracé du graphique de la fonction de répartition
plt.plot(x, cdf, color='blue')
plt.title('Fonction de répartition de la loi uniforme U(2, 5)')
plt.xlabel('x')
plt.ylabel('Fonction de répartition')
plt.grid(True)
plt.show()
```
