```python
# la densité de probabilité d'une loi exponentielle 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Paramètre de la loi exponentielle
lambda_param = 0.5  # Paramètre de taux (inverse de la moyenne)

# Valeur de x pour laquelle vous souhaitez calculer la densité de probabilité
x = 1.5

# Calcul de la densité de probabilité
pdf = expon.pdf(x, scale=1/lambda_param)
print("Densité de probabilité de la loi exponentielle avec un taux de", lambda_param, "pour x =", x, " est : ", pdf)

# Tracé du graphique de la densité de probabilité
x_values = np.linspace(0, 5, 100)  # Plage de valeurs pour l'axe x
pdf_values = expon.pdf(x_values, scale=1/lambda_param)  # Calcul de la densité de probabilité pour chaque valeur de x

plt.plot(x_values, pdf_values, color='blue')
plt.title('Densité de la loi exponentielle avec un taux de ' + str(lambda_param))
plt.xlabel('x')
plt.ylabel('Densité de probabilité')
plt.grid(True)
plt.show()
```

```python
# Paramètre de la loi exponentielle
lambda_param = 0.5  # Paramètre de taux (inverse de la moyenne)

# Plage de valeurs pour l'axe x
x_values = np.linspace(0, 5, 100)

# Calcul de la fonction de répartition pour chaque valeur de x 
cdf_values = expon.cdf(x_values, scale=1/lambda_param)

# Tracé du graphique de la fonction de répartition
plt.plot(x_values, cdf_values, color='blue')
plt.title('Fonction de répartition de la loi exponentielle avec un taux de ' + str(lambda_param))
plt.xlabel('x')
plt.ylabel('Fonction de répartition')
plt.grid(True)
plt.show()
```

```python
# Générer 5 valeurs aléatoires à partir d'une distribution exponentielle avec un paramètre lambda de 0.5
random_values = np.random.exponential(scale=1/0.5, size=5)
print(random_values)
```
