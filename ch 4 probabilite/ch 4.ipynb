```python
# Loi de Bernoulli
import numpy as np
import math as m
# Définir la probabilité de succès
p = 0.5

# Calculer la variance et l'écart-type
V = p*(1-p)
sigma = m.sqrt(V)
# Afficher les résultats
print("La probabilité de succès est p = ", p)
print("La probabilité d'èchce est q = ", 1-p)
print("Variance : ", V)
print("Espérance : ", p)
print("Écart-type : ", sigma)
```

```python
# Loi binomiale
from scipy.stats import binom
import math as m

n = 10  # nombre de pas
p = 0.5  # probabilité de réussite
k = 5  # nombre de pas en avant
prob = binom.pmf(k, n, p)
print("la probabilité qu’après 10 pas, il ait avancé de 5 m est : ", format(prob, '.2f'))
print("Variance : ", format(n*p*(1-p), '.2f'))
print("Espérance : ", n*p)
print("Écart-type : ", format(m.sqrt(n*p*(1-p)), '.2f'))
```

```python
# Loi Géométrique
from scipy.stats import geom

p = 0.7
k = 3
# Calcul de la probabilité d'obtenir une réussite après k essais
prob = geom.pmf(k, p)
print("La probabilité d'obtenir une réussite après {} essais est de {:.2f}%".format(k, prob * 100))

# Calcul de l'espérance et de la variance
esperance = geom.mean(p)
variance = geom.var(p)

print("L'espérance est : {:.2f} ".format(esperance))
print("La variance est : {:.2f} ".format(variance))
print("L'écart-type est : {:.2f} ".format(geom.std(p)))
```

```python
from scipy.stats import uniform

a = 2
b = 5

# Génère 5 nombres aléatoires suivant une loi uniforme sur l'intervalle [0,1]
rvs = uniform.rvs(size=5, loc=a, scale=b-a)

# Calcul de la probabilité que le temps de remboursement soit inférieur ou égal à 4 jours
prob = uniform.cdf(4, loc=a, scale=b-a)

# Calcule l'espérance et la variance de la distribution uniforme
esperance = uniform.mean(loc=a, scale=b-a)
variance = uniform.var(loc=a, scale=b-a)

print("La probabilité qu'un client insatisfait soit remboursé dans les 4 jours est de {:.2f}%".format(prob * 100))
print("L'espérance est de {:.2f}".format(esperance))
print("La variance est de {:.2f}".format(variance))
print("L'écart-type est de {:.2f}".format(uniform.std(loc=a, scale=b-a)))
```

```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt

a, b = 2, 10

# Création de la variable aléatoire X
X = uniform(loc=a, scale=b-a)

# Calcul de la probabilité pour X entre 3 et 8
prob = uniform.cdf(8, loc=a, scale=b-a) - uniform.cdf(3, loc=a, scale=b-a)
print("La probabilité que X soit entre 3 et 8 est de {:.2f}%".format(prob * 100))

# Calcul de l'espérance et de la variance
esperance = uniform.mean(loc=a, scale=b-a)
variance = uniform.var(loc=a, scale=b-a)
print("L'espérance de X est E(X) = {:.2f} et la variance est V(X) = {:.2f}".format(esperance, variance))

# Tracé de la densité de probabilité
x = np.linspace(a, b, 100)
y = X.pdf(x)
plt.plot(x, y)
plt.title("Densité de probabilit")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()

# Tracé de la fonction de répartition
y = X.cdf(x)
plt.plot(x, y)
plt.title("Fonction de répartition")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.show()
```

```python
from scipy.stats import hypergeom

N = 50  # taille de la population
K = 10  # nombre d'éléments de la population ayant la caractéristique recherchée
n = 20  # taille de l'échantillon

# Calcul de la probabilité d'avoir exactement 3 éléments avec la caractéristique recherchée dans l'échantillon
prob = hypergeom.pmf(3, N, K, n)
print("La probabilité d'obtenir exactement 3 éléments avec la caractéristique recherchée est de {:.2f}%".format(prob * 100))

# Calcul de l'espérance et de la variance
mean, var = hypergeom.stats(N, K, n)
print("L'espérance est E(X) = {:.2f}".format(mean))
print("La variance est V(X) = {:.2f}".format(var))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Paramètre teta de la loi exponentielle
teta = 0.5

# Calcul la probabilité X <= 5
prob = expon.cdf(5, scale=1/teta)
print("La probabilité que X soit inférieur ou égal à 5 est : {:.2f}%".format(prob * 100))

# Calcul de l'espérance et de la variance
E = expon.mean(scale=1/teta)
V = expon.var(scale=1/teta)
print("L'espérance de X est E(X) = {:.2f} et la variance de X est V(X) = {:.2f}".format(E, V))

# Calcul de la densité de probabilité et de la fonction de répartition
x = np.linspace(0, 10, 100)
pdf = expon.pdf(x, scale=1/teta)
cdf = expon.cdf(x, scale=1/teta)

# Tracé de la densité de probabilité et de la fonction de répartition
plt.plot(x, pdf, label="Densité de probabilité")
plt.plot(x, cdf, label="Fonction de répartition")
plt.legend()
plt.show()
```
