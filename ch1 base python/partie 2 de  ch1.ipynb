```python
from scipy.integrate import quad

def f(x):
    return x**2

result, error = quad(f, 0, 1)

print("Résultat de l'intégration:", result)
print("Erreur d'estimation:", error)
```

```python
from scipy.optimize import minimize

def f(x):
    return x**2 - 4*x + 4

result = minimize(f, x0=0)

print("Minimum global de la fonction:", result.x)
```

```python
from scipy.signal import lfilter
import numpy as np

# Création d'un signal sinusoïdal
t = np.linspace(0, 1, 100)
signal = np.sin(2 * np.pi * 5 * t)

# Création d'un filtre passe-bas
b = np.array([1, 2, 1]) / 4
a = np.array([1, -1.6, 0.64])

# Filtrage du signal
filtered_signal = lfilter(b, a, signal)

# Affichage des signaux
import matplotlib.pyplot as plt
plt.plot(t, signal, label="Signal original")
plt.plot(t, filtered_signal, label="Signal filtré")
plt.legend()
plt.show()

```

```python
import numpy as np

# Créer un tableau NumPy aléatoire
arr = np.random.rand(5, 5)

# Filtrer les éléments supérieurs à 0.5
filtred_arr = np.where(arr > 0.5, arr, 0)

print(filtred_arr)
```

```python
import numpy as np

# Créer un tableau NumPy aléatoire
array = np.random.rand(5, 5)

# Filtrer les éléments entre 0.2 et 0.8
filtred_array = np.where(np.logical_and(array > 0.2, array < 0.8), array, 0)

print(filtred_array)
```

```python
import numpy as np

# Créer un tableau NumPy aléatoire
array = np.random.rand(5, 5)

# Créer un masque pour filtrer les éléments inférieurs à 0.5
masked = np.ma.masked_where(array < 0.5, arr)

print(masked)
```
