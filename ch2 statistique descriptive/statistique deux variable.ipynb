```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
x = np.array([1, 1, 2, 2, 2, 3, 4, 5, 5, 5, 6, 7, 8, 8, 9, 10])
y = np.array([4, 5, 5, 3, 5, 3, 4, 4, 4, 6, 6, 5, 9, 8, 8, 7])

# Représente le nuage de points
plt.axis([0, 11, 2, 10]) 
plt.xlabel('variable x')
plt.ylabel('variable y')
plt.title('Nuage de points')
plt.grid()
plt.scatter(x, y)  

# la droite de régression
lr = linregress(x, y)  # lr reçoit cinq informations, on n'utilise que les trois premières

fit = np.polyfit(x, y, 1)
poly = np.poly1d(fit)  # On crée une fonction polynimiale à partir de a et b.
print("la droite de régression est : ")
print(poly)
poly(4)  # Affiche l'équation de la droite de régression
lx= np.linspace(0,12,100) #créer une liste de 100 valeurs dans l'intervalle[0,30]
plt.plot(lx,poly(lx)) # DEssine la droite de régression
plt.show() # Commande qui affiche la représentation
```

```python
import numpy as np
from scipy.stats import linregress

x = np.array([1, 1, 2, 2, 2, 3, 4, 5, 5, 5, 6, 7, 8, 8, 9, 10])
y = np.array([4, 5, 5, 3, 5, 3, 4, 4, 4, 6, 6, 5, 9, 8, 8, 7])

print("coefficient de corrélation est Cov(X,Y) = ", format(np.cov(x, y)[0][1], '.2f')) # afficher coefficient de corrélation est Cov(X,Y) =  3.98
print("coefficient de corrélation est r(X,Y) = ", format(linregress(x, y)[2], '.2f')) # afficher coefficient de corrélation est r(X,Y) =  0.75
```

```python
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Données d'exemple
x =  np.array([1, 1, 2, 2, 2, 3, 4, 5, 5, 5, 6, 7, 8, 8, 9, 10])
y = np.array([4, 5, 5, 3, 5, 3, 4, 4, 4, 6, 6, 5, 9, 8, 8, 7])

# Ajustement de la droite de Mayer
slope, intercept, r_value, p_value, std_err = linregress(x, y)
y_fit = slope * x + intercept

# Affichage des résultats
print("Coefficient directeur de la droite de Mayer :", slope)
print("Ordonnée à l'origine de la droite de Mayer :", intercept)

# Tracé de la droite de Mayer et des données
plt.scatter(x, y)
plt.plot(x, y_fit, 'r-')
plt.xlabel('Variable x')
plt.ylabel('Variable y')
plt.title('la droite de Mayer')
plt.show()
```
