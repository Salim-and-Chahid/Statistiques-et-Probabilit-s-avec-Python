```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
data = pd.read_excel('exercice.xlsx',sheet_name = 'donnees')
data
```

```python
sns.lmplot(x = 'Chambres', y = 'Prix', data = data, ci = None);
```

```python
X = data['Chambres']
Y = data['Prix']
model = stats.linregress(X, Y)
nb_Chambres = 7
estimation = model.slope*nb_Chambres + model.intercept
estimation
```
