```python
# importer les bibliotique necessaires
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# charger les données à partir d'un fichiers excel
data=pd.read_excel("regression_multiple.xlsx")
data #pour afficher les donnees
```

```python
# la correlation entre les vqriqbles
corr=data.corr()
sns.heatmap(corr, annot=True)
```

```python
# représentation graphique de corrélation en utilisant seaborn 
sns.pairplot(data)
```

```python
# séparer les données 
X=data[['M2','Année']]
Y=data[['Prix']]
```

```python
# créer le modèle de regression lineaire multiple
model=LinearRegression()
model.fit(X,Y)
Intercept=model.intercept_
Coefficients=model.coef_
Score=model.score(X,Y)
```

```python
# afficher les élements
print(Intercept, Coefficients, Score)
```

```python
new_data=pd.DataFrame({'M2':[100],'Annee':[2030]})
prediction=model.predict(new_data)
prediction
```
