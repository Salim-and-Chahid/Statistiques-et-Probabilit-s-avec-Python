```python
import numpy as np
import matplotlib.pyplot as plt

# Données d'exemple
x = np.array([8, 10, 12, 14, 16, 18])
y = np.array([40, 55, 55, 70, 75, 95])

# Calcul de la droite de Mayer
cov_xy = np.cov(x, y)[0][1]
var_x = np.var(x)
b = cov_xy / var_x
a = np.mean(y) - b * np.mean(x)

# Tracé du diagramme de dispersion et de la droite de Mayer
plt.scatter(x, y)
plt.plot(x, a + b * x, color='red')
plt.xlabel('Variable x')
plt.ylabel('Variable y')
plt.title('Droite d’ajustement de Mayer')
plt.show()


```
